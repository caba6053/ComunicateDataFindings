## Univariate Exploration

**What is the most common credit grade in the dataset?**

```python
freq = df['CreditGrade'].value_counts()
gen_order = freq.index
base_color = sb.color_palette()[0]
sb.countplot(data=df, x='CreditGrade', color=base_color, order=gen_order);
```

**C is the most common, followed by D then B**

**What is the most common income range?**

```python
type_order = df['IncomeRange'].value_counts().index
sb.countplot(data=df, y='IncomeRange', color=base_color, order=type_order);
```

**25k-49k is the most common with 50k-74k as the next common**

**What kind of loans are in this dataset? (Past, present, future)**

```python
type_order = df['LoanStatus'].value_counts().index
sb.countplot(data=df, y='LoanStatus', color=base_color, order=type_order);
```

**Mainly current loans with some completed as well and little that are past due**

**Is the majority of people with loans employed?**

```python
type_order = df['EmploymentStatus'].value_counts().index
sb.countplot(data=df, y='EmploymentStatus', color=base_color, order=type_order);
```

**Yes, the majority of people are employed. However, some people are just listed as 'Employed' while others are listed as how often they work.**

**Do people with higher credit scores have loans?**

```python
type_order = df['CreditScoreRangeLower'].value_counts().index
sb.countplot(data=df, y='CreditScoreRangeLower', color=base_color, order=type_order);
```

```python
type_order = df['CreditScoreRangeUpper'].value_counts().index
sb.countplot(data=df, y='CreditScoreRangeUpper', color=base_color, order=type_order);
```

**Typically it is people between 640-739.**

**What are the interest rates for loans?** 

```python
plt.hist(data = df, x = 'BorrowerRate', bins = 15);
```

**The graph is skewed right and the range is 0-0.375.**

**What are the borrower's annual percentage rate?**

```python
plt.hist(data = df, x = 'BorrowerAPR', bins = 15);
```

**The borrower APR follows the same coorelation as the BorrowerRate.**

**What are people using their loans for?**

```python
sorted_counts = df['ListingCategory (numeric)'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False);
plt.axis('square');
```

**Debt consolidation, home improvement, and business**

## Bivariate Exploration

**How well does APR and interest rate coorelate with each other?**

```python
sb.regplot(data = df, x = 'BorrowerAPR', y = 'BorrowerRate');
plt.xlabel('APR')
plt.ylabel('Interest Rate')
```

**There is a strong coorelation between the APR and interest rate as aspected. <br> There is one outlier that has a lower APR compared to its interest rate. <br> There does not seem to be many values with an APR and interest rate equal or greater than 0.4.**

**Does interest rate decrease with income?**

```python
base_color = sb.color_palette()[0]
sb.violinplot(data=df, x='IncomeRange', y='BorrowerRate', color=base_color, innner=None)
plt.xticks(rotation=25);
```

**Yes, the average interest rate is highest with unemployed and lowest with 100k+ range. <br> The 'Not displayed' and '$0' seem to be misleading information. <br> The range of interest rate is very close for all incomes.**

**Does APR look similiar to interest rate in terms of income range?**

```python
base_color = sb.color_palette()[0]
sb.violinplot(data=df, x='IncomeRange', y='BorrowerAPR', color=base_color, innner=None)
plt.xticks(rotation=25);
```

**Yes, this plot is almost identical to the one before it.**

## Multivariate Exploration

**Does credit score, interest rate, APR, and homeownership coorelate?**

```python
g = sb.FacetGrid(data = df, hue = 'IsBorrowerHomeowner', 
                 hue_order = [False, True], size=4, aspect=1.5)
g.map(sb.regplot, 'CreditScoreRangeUpper', 'BorrowerRate', x_jitter = 0.4, fit_reg = False)
g.add_legend()
plt.xlabel('Credit Score Upper')
plt.ylabel('Interest Rate');
```

```python
g = sb.FacetGrid(data = df, hue = 'IsBorrowerHomeowner', 
                 hue_order = [False, True], size=4, aspect=1.5)
g.map(sb.regplot, 'CreditScoreRangeUpper', 'BorrowerAPR', x_jitter = 0.4, fit_reg = False)
g.add_legend()
plt.xlabel('Credit Score Upper')
plt.ylabel('APR');
```

**Interest rate and APR have the same coorelation so they can be interchnaged. <br> The higher the credit score the more likely a person is a homeowner. <br> Interest rate and credit score do not seem to have a pattern with each other. <br> Interest rate and homeownership do not have a coorelation either.**

**How does avaliable credit, credit score, and credit grade coorelate with each other?**

```python
g = sb.FacetGrid(data = df, hue = 'CreditGrade', size=4, aspect=1.5)
g.map(sb.regplot, 'CreditScoreRangeUpper', 'AvailableBankcardCredit', x_jitter = 0.4, fit_reg = False)
g.add_legend()
plt.xlabel('Credit Score')
plt.ylabel('Available Credit');
```

**There is a positive coorelation between credit score and available credit. <br>
There is also a positive coorelation between credit grade and credit score. <br>
Lastly, there is a positive coorelation between avialable credit and credit grade.**

**How does income, interest rate, and bank credit compare?**

```python
g = sb.FacetGrid(data = df, hue = 'IncomeRange', size=4, aspect=1.5)
g.map(sb.regplot, 'BorrowerRate', 'AvailableBankcardCredit', x_jitter = 0.4, fit_reg = False)
g.add_legend()
```

**There does not seem to be a coorelation between interest rate and avialable bank credit. <br>
There also no coorelation between interest rate and income. <br>
There is a slight positive coorealtion between income and avlailable bank credit.**

**How does current credit lines, avialable bankcard credit, income, and credit grade coorelate?**

```python
g = sb.FacetGrid(data = df, col = 'IncomeRange', row = 'CreditGrade', size = 2.5,
                margin_titles = True)
g.map(plt.scatter, 'CurrentCreditLines', 'AvailableBankcardCredit')
```

**The highest avialable credit is people in the highest income range with the highest credit grade in the mid to high current credit lines. <br>
The only time avialable credit increases with credit card line is when the credit grade score is a 'AA'. <br>
Income plays a role in avialable credit only when the credit grade score is a 'B' or higher otherwise the amount of credit lines and income does not have an effect on avialable credit.**

## Observations:

**There are some positive coorelations I am not suprised about. These include: interest rate and APR, credit score and available credit,
credit grade and credit score, available credit and credit grade, and credit score and being a homeowner. <br>
If someone has a high credit score they are going to have more available credit and a higher credit grade. <br>
If someone has a high credit score they are more likely to be a homeowner.**

**There are some negative coorelations I am not surprised about. These include: income range and interest rate, and 
    income range and APR. <br>
    Since interest rate and APR both have a strong positive coorelation with each other they are practically interchangable.
    As income range increases, there is a lower average of interest rate and APR.**

**I am surprised that available bank credit and interest rate has no coorelation with each other considering interest rate coorelates with income.
Interest rate and credit score do not coorelate as well and since available bank credit coorelates with credit score these observations do make sense.
<br>
The most interesting observation is that available bank credit has a more significant change with credit grade rather than amount of credit lines.
Income does not coorelate with available bank credit and number of credit lines if the credit grade is lower than a "B". 
Credit grade has a larger role in avilable bank credit than income, but the amount of credit lines so help as well.**
